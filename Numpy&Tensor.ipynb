# Numpy
- **Fast computation**: NumPy arrays store elements of the same data type, so there is no need for extra type checking during operations.
- **Memory efficiency**: NumPy does repeated calculations in one go, so it uses less memory and runs quicker than lists.
- **Effect**: The bigger the data, the bigger the benefit.

```python
import numpy as np
```

## Basic Code (1)

```python
a = np.array([1, 2, 3])   # create a 1D array
print(type(a))             # output: "<type 'numpy.ndarray'>"
print(a.shape)             # output: "(3,)"
print(a.ndim)              # output: "1"
print(a[0], a[1], a[2])    # output: "1 2 3"
a[0] = 5                   # change an element
print(a)                   # output: "[5, 2, 3]"

b = np.array([[1,2,3],[4,5,6]])  # create a 2D array (rank 2)
print(b.shape)                    # output: "(2, 3)"
print(b.ndim)                     # output: "2"
print(b[0, 0], b[0, 1], b[1, 0])  # output: "1 2 4"
```

## Basic Code (2)

```python
import numpy as np

a = np.zeros((2,2))       # create an array filled with zeros
print(a)                  # output "[[0. 0.]
                          #          [0. 0.]]"

b = np.ones((1,2))        # create an array filled with ones
print(b)                  # output "[[1. 1.]]"

c = np.full((2,2), 7)     # create an array filled with a constant value
print(c)                  # output "[[7. 7.]
                          #          [7. 7.]]"

d = np.eye(2)             # create a 2x2 identity matrix
print(d)                  # output "[[1. 0.]
                          #          [0. 1.]]"

e = np.random.random((2,2))   # create an array filled with random values
print(e)                      # output random values, e.g. "[[0.91940167 0.08143941]
                              #                           [0.68744134 0.87236687]]"
```

# Tensors
- **GPU acceleration**: Tensors can run on GPUs/TPUs for much faster computation.
- **Automatic differentiation**: Tensors keep track of operations, so gradients can be computed automatically.
- **NumPy and Tensors are almost the same!**

image.png

```python
import torch
```

## Tensor (1)

```python
import torch
import numpy as np

# Create tensors
t1 = torch.FloatTensor([0, 1, 2, 3, 4, 5, 6])
t2 = torch.tensor(np.arange(7))   # Same result

# Check shapes
print(t1.shape, t2.shape)         # torch.Size([7]) torch.Size([7])

# Dimension, shape, size
print(t1.dim())                   # output: 1
print(t1.shape)                   # output: torch.Size([7])
print(t1.size())                  # output: torch.Size([7])

# Indexing and slicing
print(t1[:2], t1[3:])             # output: tensor([0., 1.]) tensor([3., 4., 5., 6.])
```

## Tensor (2)

```python
# Matrix addition
m1 = torch.FloatTensor([[3, 3]])
m2 = torch.FloatTensor([[2, 2]])
print(m1 + m2)               # output: [[5, 5]]

# Vector + scalar (broadcasting)
m1 = torch.FloatTensor([[1, 2]])
m2 = torch.FloatTensor([3])  # 3 -> [[3, 3]]
# m2 = torch.FloatTensor(3)  # Error
print(m1 + m2)               # output: [[4, 5]]

# 2x1 Vector + 1x2 Vector (broadcasting)
m1 = torch.FloatTensor([[1, 2]])
m2 = torch.FloatTensor([[3], [4]])
print(m1 + m2)
# output:
# [[4, 5]
#  [5, 6]]
```

## Mul vs MatMul

```python
print('-------------')
print('Mul vs Matmul')
print('-------------')

# Matrix multiplication (matmul)
m1 = torch.FloatTensor([[1, 2], [3, 4]])
m2 = torch.FloatTensor([[1], [2]])
print('Shape of Matrix 1: ', m1.shape)   # 2 x 2
print('Shape of Matrix 2: ', m2.shape)   # 2 x 1
print(m1.matmul(m2))                     # result: 2 x 1

# Element-wise multiplication (mul)
m1 = torch.FloatTensor([[1, 2], [3, 4]])
m2 = torch.FloatTensor([[1], [2]])
print('Shape of Matrix 1: ', m1.shape)   # 2 x 2
print('Shape of Matrix 2: ', m2.shape)   # 2 x 1
print(m1 * m2)                           # result: 2 x 2
print(m1.mul(m2))                        # same as above
```

## Vew (Numpy: Reshape)

```python
# Create a NumPy array with shape (2, 2, 3)
a = np.arange(12).reshape(-1, 2, 3)
print(a.shape)                  # (2, 2, 3)

# Convert to FloatTensor
b = torch.FloatTensor(a)
print(b.shape)             # torch.Size([2, 2, 3])

# Reshape using view
print(b.view([-1, 3]))     # reshape to (X, 3)
```

## Linear Regression

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

# Training data
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [3], [4]])

# Initialize model parameters
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# Set optimizer
optimizer = optim.SGD([W, b], lr=0.01)

# Number of epochs
nb_epochs = 1000
```

```python
for epoch in range(nb_epochs + 1):

    # Compute hypothesis H(x)
    hypothesis = x_train * W + b

    # Compute cost (MSE)
    cost = torch.mean((hypothesis - y_train) ** 2)

    # Update parameters using gradient descent
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    # Print log every 100 epochs
    if epoch % 100 == 0:
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f}, Cost: {:.6f}'.format(
            epoch, nb_epochs, W.item(), b.item(), cost.item()
        ))
```
